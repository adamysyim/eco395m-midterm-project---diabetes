```python
import pandas as pd
```

```python
# Read in the dataset
df = pd.read_sas("../data/llcp2017.xpt", encoding='utf-8')
```

```python
# check number of rows and columns
df.shape
```

```python
# check that the first 5 rows with full display of columns
pd.set_option('display.max_columns', 300)
df.head()
```

```python
# select columns for analysis
df_sel = df[['DIABETE3','_PHYS14D','PA1VIGM_','_PACAT1','_PA150R2','_PA300R2','_PA30021','_PAINDX1','_PASTRNG',
                    '_PAREC1','_VEGLT1A','_FRTLT1A','POTADA1_','FRNCHDA_','GRENDA1_','FRUTDA2_','FTJUDA2_','_RFDRHV5',
                    '_RFBING5','_CURECIG','_ECIGSTS','_RFSMOK3','_SMOKER3','_INCOMG','_EDUCAG','_BMI5','_BMI5CAT','_RFBMI5',
                    '_MICHD','CVDSTRK3','_RFCHOL1','TOLDHI2','_RFHYPE5','_MENT14D','MENTHLTH','SEX','_AGEG5YR','_AGE_G','_AGE65YR']]
```

```python
df_sel.head()
```

```python
# drop missing values
df_sel = df_sel.dropna()
df_sel['DIABETE3'].value_counts()
```

```python
## DIABETE3 - check,replace,remove
 # 0 is for no diabetes or only during pregnancy, 1 is for pre-diabetes or borderline diabetes, 2 is for yes diabetes
 # remove 7(don't know) and 9(refused)
df_sel['DIABETE3'] = df_sel['DIABETE3'].astype(int)
df_sel['DIABETE3'].replace({2:0,3:0,1:2,4:1},inplace=True)
df_sel['DIABETE3'].value_counts()
```

```python
df_sel = df_sel[df_sel['DIABETE3'] != 7]
df_sel = df_sel[df_sel['DIABETE3'] != 9]
df_sel['DIABETE3'].value_counts()
```

```python
## _PHYS14D - check,replace,remove
 # removing 9(don't know)
 # change order and scale 
  # 1 >> 0 (Zero days when physical health not good)
  # 2 >> 1 (1-13 days when physical health not good)
  # 3 >> 2(14+ days when physical health not good)
  # Inactive 4 >> 0 
df_sel=df_sel[df_sel['_PHYS14D'] != 9]
df_sel['_PHYS14D'].replace({1:0,2:1,3:2},inplace=True)
df_sel['_PHYS14D'] = df_sel['_PHYS14D'].astype(int)
df_sel['_PHYS14D'].value_counts()
```

```python
## PA1VIGM_ - check,replace,remove
df_sel['PA1VIGM_'] = df_sel['PA1VIGM_'].round(0)
df_sel['PA1VIGM_'].value_counts()
```

```python
## _PACAT1 - check,replace,remove
 # removing 9(don't know)
 # change order and scale (inactive(0) to highly active(3)) 
  # Highly Active 1 >> 4
  # Active 2 >> 3
  # Insufficiently Active 3 >> 1
  # Inactive 4 >> 0 
df_sel=df_sel[df_sel['_PACAT1'] != 9]
df_sel['_PACAT1'] = df_sel['_PACAT1'].astype(int)
df_sel['_PACAT1'].replace({1:3,2:2,3:1,4:0},inplace=True)
df_sel['_PACAT1'].value_counts()
```

```python
## _PA150R2 - check,replace,remove
# change order and scale 
 # 1 >> 2
 # 2 >> 1
 # 3 >> 0
df_sel['_PA150R2'].replace({1:2,2:1,3:0},inplace=True)
df_sel['_PA150R2'] = df_sel['_PA150R2'].astype(int)
df_sel['_PA150R2'].value_counts()
```

```python
## _PA300R2 - check,replace,remove
 # change order and scale 
  # 1 >> 2
  # 2 >> 1
  # 3 >> 0
df_sel['_PA300R2'].replace({1:2,2:1,3:0},inplace=True)   
df_sel['_PA300R2'] = df_sel['_PA300R2'].astype(int)
df_sel['_PA300R2'].value_counts()
```

```python
## _PA30021 - check,replace,remove
# change order and scale 
 # 1 >> 1 (301+minutes)
 # 2 >> 0 (0-300minutes)
df_sel['_PA30021'].replace({1:1,2:0},inplace=True)  
df_sel['_PA30021'] = df_sel['_PA30021'].astype(int)
df_sel['_PA30021'].value_counts()
```

```python
## _PAINDX1 - check,replace,remove
 # change order and scale 
  # 1 >> 1 (Meet Aerobic Recommendations)
  # 2 >> 0 (Did Not Meet Aerobic Recommendations)
df_sel['_PAINDX1'].replace({1:1,2:0},inplace=True)  
df_sel['_PAINDX1'] = df_sel['_PAINDX1'].astype(int)
df_sel['_PAINDX1'].value_counts()
```

```python
## _PASTRNG - check,replace,remove
 # removing 9(don't know)
 # change order and scale 
  # 1 >> 1 (Meet muscle strengthening recommendations)
  # 2 >> 0 (Did not meet muscle strengthening recommendations)
df_sel=df_sel[df_sel['_PASTRNG'] != 9]
df_sel['_PASTRNG'].replace({1:1,2:0},inplace=True)
df_sel['_PASTRNG'] = df_sel['_PASTRNG'].astype(int)
df_sel['_PASTRNG'].value_counts()
```

```python
## _PAREC1 - check,replace,remove
 # change order, scale, merge 
  # 1 >> 2 (Met Both Guidelines)
  # 2 and 3 >> 1 (Met Aerobic Guidelines Only or Met Strengthening Guidelines Only)
  # 4 >> 0 (Did not meet Either Guideline)
df_sel['_PAREC1'].replace({1:2,2:1,3:1,4:0},inplace=True)
df_sel['_PAREC1'] = df_sel['_PAREC1'].astype(int)
df_sel['_PAREC1'].value_counts()
```

```python
## _VEGLT1A - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 1 (Consumed vegetables one or more times per day)
  # 2 >> 0 (Consumed vegetables less than one time per day)
df_sel=df_sel[df_sel['_VEGLT1A'] != 9]
df_sel['_VEGLT1A'].replace({1:1,2:0},inplace=True)
df_sel['_VEGLT1A'] = df_sel['_VEGLT1A'].astype(int)
df_sel['_VEGLT1A'].value_counts()
```

```python
## _FRTLT1A - check,replace,remove
 # change order, scale
  # 1 >> 1 (Consumed fruit one or more times per day)
  # 2 >> 0 (Consumed fruit less than one time per day)
df_sel['_FRTLT1A'].replace({1:1,2:0},inplace=True)
df_sel['_FRTLT1A'] = df_sel['_FRTLT1A'].astype(int)
df_sel['_FRTLT1A'].value_counts()
```

```python
## POTADA1_ - check,replace,remove
df_sel['POTADA1_'] = df_sel['POTADA1_'].round(0)
df_sel['POTADA1_'].value_counts()
```

```python
## FRNCHDA_ - check,replace,remove
df_sel['FRNCHDA_'] = df_sel['FRNCHDA_'].round(0)
df_sel['FRNCHDA_'].value_counts()
```

```python
## GRENDA1_ - check,replace,remove
 # removing extreme value
df_sel = df_sel[df_sel['GRENDA1_'] >= 1]    
df_sel['GRENDA1_'].value_counts()
```

```python
## FRUTDA2_ - check,replace,remove
df_sel = df_sel[df_sel['FRUTDA2_'] >= 1] 
df_sel['FRUTDA2_'].value_counts()
```

```python
## FTJUDA2_ - check,replace,remove
df_sel['FTJUDA2_'] = df_sel['FTJUDA2_'].round(0)
df_sel['FTJUDA2_'].value_counts()
```

```python
## _RFDRHV5 - check,replace,remove
 # removing 9(don't know)
df_sel=df_sel[df_sel['_RFDRHV5'] != 9]
df_sel['_RFDRHV5'] = df_sel['_RFDRHV5'].astype(int)
df_sel['_RFDRHV5'].value_counts()
```

```python
## _RFBING5 - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 0 (No)
  # 2 >> 1 (Yes)
df_sel=df_sel[df_sel['_RFBING5'] != 9]
df_sel['_RFBING5'].replace({1:0,2:1},inplace=True)
df_sel['_RFBING5'] = df_sel['_RFBING5'].astype(int)
df_sel['_RFBING5'].value_counts()
```

```python
## _CURECIG - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 1 (Current E-cigarette user)
  # 2 >> 0 (Not currently using E-cigarettes)
df_sel=df_sel[df_sel['_CURECIG'] != 9]
df_sel['_CURECIG'].replace({1:1,2:0},inplace=True)
df_sel['_CURECIG'] = df_sel['_CURECIG'].astype(int)
df_sel['_CURECIG'].value_counts()
```

```python
## _ECIGSTS - check,replace,remove
 # change order, scale
  # 1 >> 3 (Current E-cigarette user - uses every day)
  # 2 >> 2 (Current E-cigarette user - uses some days)
  # 3 >> 1 (Former E-cigarette user)
  # 4 >> 0 (Never used E-cigarettes)
df_sel['_ECIGSTS'].replace({1:3,2:2,3:1,4:0},inplace=True)   
df_sel['_ECIGSTS'] = df_sel['_ECIGSTS'].astype(int)
df_sel['_ECIGSTS'].value_counts()
```

```python
##  _RFSMOK3 - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 0 (No)
  # 2 >> 1 (Yes)
df_sel=df_sel[df_sel['_RFSMOK3'] != 9]
df_sel['_RFSMOK3'].replace({1:0,2:1},inplace=True) 
df_sel['_RFSMOK3'] = df_sel['_RFSMOK3'].astype(int)
df_sel['_RFSMOK3'].value_counts()
```

```python
##  _SMOKER3 - check,replace,remove
 # change order, scale
  # 1 >> 3 (Current smoker - now smokes every day)
  # 2 >> 2 (Current smoker - now smokes some days)
  # 3 >> 1 (Former smoker)
  # 4 >> 0 (Never smoked)
df_sel['_SMOKER3'].replace({4:0,3:1,2:2,1:3},inplace=True)
df_sel['_SMOKER3'] = df_sel['_SMOKER3'].astype(int)
df_sel['_SMOKER3'].value_counts()
```

```python
##  _INCOMG - check,replace,remove
 # removing 9(don't know)
 # change order, scale

df_sel=df_sel[df_sel['_INCOMG'] != 9]
df_sel['_INCOMG'] = df_sel['_INCOMG'].astype(int)
df_sel['_INCOMG'].value_counts()
```

```python
##  _EDUCAG - check,replace,remove
 # removing 9(don't know)
df_sel=df_sel[df_sel['_EDUCAG'] != 9]
df_sel['_EDUCAG'] = df_sel['_EDUCAG'].astype(int)
df_sel['_EDUCAG'].value_counts()
```

```python
##  _BMI5 - check,replace,remove
 # THe original value is BMI * 100. Therefore, divide every value with 100
df_sel['_BMI5'] = df_sel['_BMI5'].div(100).round(0)
df_sel['_BMI5'] = df_sel['_BMI5'].astype(int)
df_sel['_BMI5'].value_counts()
```

```python
## _RFBMI5 - check,replace,remove
# removing 9(don't know)
 # change order, scale
  # 1 >> 0 (No)
  # 2 >> 1 (Yes - obese)
df_sel=df_sel[df_sel['_RFBMI5'] != 9]
df_sel['_RFBMI5'].replace({1:0,2:1},inplace=True)
df_sel['_RFBMI5'] = df_sel['_RFBMI5'].astype(int)
df_sel['_RFBMI5'].value_counts()
```

```python
## _BMI5CAT - check,replace,remove
df_sel['_BMI5CAT'] = df_sel['_BMI5CAT'].astype(int)
df_sel['_BMI5CAT'].value_counts()
```

```python
## _MICHD - check,replace,remove
 # change order, scale
  # 1 >> 1 (Reported having MI or CHD)
  # 2 >> 0 (Did not report having MI or CHD)
df_sel['_MICHD'].replace({1:1,2:0},inplace=True)
df_sel['_MICHD'] = df_sel['_MICHD'].astype(int)
df_sel['_MICHD'].value_counts()
```

```python
## CVDSTRK3 - check,replace,remove
 # removing 7(don't know) 9(refused)
 # change order, scale
  # 1 >> 1 (Yes)
  # 2 >> 0 (No)

df_sel=df_sel[df_sel['CVDSTRK3'] != 7]
df_sel=df_sel[df_sel['CVDSTRK3'] != 9]
df_sel['CVDSTRK3'].replace({1:1,2:0},inplace=True)
df_sel['CVDSTRK3'] = df_sel['CVDSTRK3'].astype(int)
df_sel['CVDSTRK3'].value_counts()
```

```python
## _RFCHOL1 - check,replace,remove
 # change order, scale
  # 1 >> 0 (No)
  # 2 >> 1 (Yes)
df_sel['_RFCHOL1'].replace({1:0,2:1},inplace=True)  
df_sel['_RFCHOL1'] = df_sel['_RFCHOL1'].astype(int)
df_sel['_RFCHOL1'].value_counts()
```

```python
## TOLDHI2 - check,replace,remove
 # change order, scale
  # 1 >> 1 (Yes)
  # 2 >> 0 (No)
df_sel['TOLDHI2'].replace({1:1,2:0},inplace=True)  
df_sel['TOLDHI2'] = df_sel['TOLDHI2'].astype(int)
df_sel['TOLDHI2'].value_counts()
```

```python
## _RFHYPE5 - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 0 (No)
  # 2 >> 1 (Yes)
df_sel=df_sel[df_sel['_RFHYPE5'] != 9]
df_sel['_RFHYPE5'].replace({1:0,2:1},inplace=True) 
df_sel['_RFHYPE5'] = df_sel['_RFHYPE5'].astype(int)
df_sel['_RFHYPE5'].value_counts()
```

```python
## _MENT14D - check,replace,remove
 # removing 9(don't know)
 # change order, scale
  # 1 >> 2 (Zero days when mental health not good)
  # 2 >> 1 (1-13 days when mental health not good)
  # 3 >> 0 (14+ days when mental health not good)
df_sel=df_sel[df_sel['_MENT14D'] != 9]
df_sel['_MENT14D'].replace({1:2,2:1,3:0},inplace=True) 
df_sel['_MENT14D'] = df_sel['_MENT14D'].astype(int)
df_sel['_MENT14D'].value_counts()
```

```python
## MENTHLTH - check,replace,remove
 # change 88 to 0 (no bad mental health days)
df_sel['MENTHLTH'] = df_sel['MENTHLTH'].replace({88:0})    
df_sel['MENTHLTH'].value_counts()
```

```python
## SEX - check,replace,remove
df_sel['SEX'] = df_sel['SEX'].astype(int)
df_sel['SEX'].value_counts()
```

```python
## _AGEG5YR - check,replace,remove
df_sel['_AGEG5YR'] = df_sel['_AGEG5YR'].astype(int)
df_sel['_AGEG5YR'].value_counts()
```

```python
## _AGE_G - check,replace,remove
df_sel['_AGE_G'] = df_sel['_AGE_G'].astype(int)
df_sel['_AGE_G'].value_counts()
```

```python
## _AGE65YR - check,replace,remove
df_sel['_AGE65YR'] = df_sel['_AGE65YR'].astype(int)
df_sel['_AGE65YR'].value_counts()
```
